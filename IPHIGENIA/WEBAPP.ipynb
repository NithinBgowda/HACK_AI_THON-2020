```
#install and import necessary packages
!pip install flask-ngrok
import tensorflow as tf
import cv2 as cv
import flask
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing import image
from tensorflow.keras.preprocessing import image
import os
from flask_ngrok import run_with_ngrok
from flask import Flask, render_template, request 

#Load the model
model = tf.keras.models.load_model('/content/drive/MyDrive/webapp/model/cov_pre_mod.h5') 


app = flask.Flask(__name__, template_folder='/content/drive/My Drive/webapp/templates')
run_with_ngrok(app) 

app.config['UPLOAD_FOLDER'] = '/content/drive/My Drive/webapp/uploaded/image'
@app.route('/', methods=['GET', 'POST'])
def main():
    if flask.request.method == 'GET':
        return(flask.render_template('main.html'))

    if flask.request.method == 'POST':
        images1 = []

        vals = ['COVID-19', 'NORMAL', 'VIRAL PNEUMONIA']

        image_size = 256
        input_variables = flask.request.files['avatar']
        print(input_variables.filename)
        FN = input_variables.filename
        input_variables.save(os.path.join(app.config['UPLOAD_FOLDER'], FN))
   
        IMA = os.path.join('/content/drive/My Drive/webapp/uploaded/image/', FN)

        image1_read = cv.imread(IMA, 0)
        image1_resized = cv.resize(image1_read,(image_size ,image_size),cv.IMREAD_GRAYSCALE)/255.0
        images1.append(image1_resized)
        images1 = np.array(images1)
        images1 = np.expand_dims(images1, axis = 3)

        prediction = model.predict(images1)
        print(prediction) 
        r = str(vals[np.argmax(prediction)])

        print(r)
        
        return flask.render_template('pred.html', ss=r)

if __name__ == '__main__':
    app.run()
```
